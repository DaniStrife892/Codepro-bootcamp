```python
Contrasena = 90
Usuario = int(input('Ingrese contrasena:'))

if Usuario == Contrasena:
    print("¡Contraseña correcta!")
else:
    print("Contraseña incorrecta")
while Usuario != Contrasena:
    print("Contraseña incorrecta. Intenta de nuevo.")
    Usuario = int(input('Ingrese contrasena: '))

print("¡Contraseña correcta!")
```
